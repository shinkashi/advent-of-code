```typescript
const makeOps = (len: number) => {
    const output = [];
    for (let i = 0; i < 2**len; i++) {
        const res = [];
        for (let j = 0; j < len; j++) {
            res.push(i & 1<<j ? "*" : "+");
        }
        output.push(res);
    }
    return output;
}

makeOps(1)
```

```typescript
const calcOps = (_nums, ops, max): boolean => {
    const nums = [..._nums];
    let x = nums.shift();
    // if (x > max) return false;
    while (nums.length > 0) {
        const op = ops.shift();
        const n = nums.shift();
        // console.log({x, op, n, ops, nums})
        if (op === "*") x = x * n;
        if (op === "+") x = x + n;
        if (x > max) return false;
    }
    return (x === max);
}

calcOps([81,40,27], ["*", "+" ], 3267)
calcOps([81,40,27], ["+", "*" ], 3267)
```

```typescript
const makeOpsCat = (len: number) => {
    const output = [];
    for (let i = 0; i < 3**len; i++) {
        const iRadix = ("0".repeat(len) + i.toString(3)).slice(-len)
        // console.log({iRadix});
        const res = [];
        for (let j = 0; j < len; j++) {
            res.push({"0": "+", "1": "*", "2": "||"}[iRadix[j]]);
        }
        output.push(res);
    }
    return output;
}

makeOpsCat(2)
```

```typescript
const calcOpsCat = (_nums, ops, max): boolean => {
    const nums = [..._nums];
    let x = nums.shift();
    // if (x > max) return false;
    while (nums.length > 0) {
        const op = ops.shift();
        const n = nums.shift();
        // console.log({x, op, n, ops, nums})
        if (op === "*") x = x * n;
        if (op === "+") x = x + n;
        if (op === "||") x = Number(x.toString() + n.toString());
        if (x > max) return false;
    }
    return (x === max);
}

// calcOpsCat([81,40,27], ["*", "+" ], 3267)
// calcOpsCat([81,40,27], ["+", "*" ], 3267)
//calcOpsCat([15, 6], ["||"], 156)
calcOpsCat([6, 8,6, 15], ["*", "||", "*"], 7290)

///7290: 6 8 6 15 can be made true using 6 * 8 || 6 * 15.
///192: 17 8 14 can be made true using 17 || 8 + 14.
```
