```typescript
const field = input1.split('\n').map(l => Array.from(l))

const ymax = field.length;
const xmax = field[0].length;

console.log({ymax, xmax})
```

```typescript
class Position {
    constructor(public y: number, public x: number) {};

    getAntinodes(other: Position): Position[] {
        const antinodes = []

        const ys = [this.y, other.y]; // .sort();
        const ya = ys[0] - (ys[1] - ys[0]);
        const yb = ys[1] + (ys[1] - ys[0]);
        
        const xs = [this.x, other.x]; // .sort();
        const xa = xs[0] - (xs[1] - xs[0]);
        const xb = xs[1] + (xs[1] - xs[0]);

        if (0 <= xa && xa < xmax && 0 <= ya && ya < ymax) {
            antinodes.push(new Position(ya, xa))
        }
        if (0 <= xb && xb < xmax && 0 <= yb && yb < ymax) {
            antinodes.push(new Position(yb, xb))
        }

        return antinodes;
    }
}

const p1 = new Position(1, 8)
const p2 = new Position(2, 5)
p1.getAntinodes(p2)








```

```typescript

class Positions {
    s: Set
    
    constructor() { this.s = new Set(); }

    locKey(p: Position) { return `${p.y},${p.x}` }

    put(p: Position) { this.s.add(this.locKey(p)) }

    get(p: Position) { return this.s.has(this.locKey(p)) }

    *list() {
        for (const loc of this.s) {
            const locNum = loc.split(",").map(z => Number(z))
            yield new Position(locNum[0], locNum[1])
        }
    }

    print() {
        let result = "";
        for (let y = 0; y < ymax; y++) {
            for (let x = 0; x < xmax; x++) {
                if (this.get(new Position(y, x))) {
                    result += "#"
                } else {
                    result += "."
                }
            }
            result += "\n"
        }
        return result;
    }
}

const ps = new Positions();
ps.put(new Position(1,2))
ps.put(new Position(3,4))

for (const p of ps.list()) {
    console.log(p)
}

ps.print()
```

```typescript

// const frequencies: {[frequency: string]: Set} = {};

const globalAntinodes = new Positions();
const antennas = new Positions();

const place = (p: Position, antennas: Positions) => {
    for (const antenna of antennas.list()) {
        const antinodes = p.getAntinodes(antenna)
        for (const antinode of antinodes) globalAntinodes.put(antinode);
    }
    antennas.put(p);
}


// place(new Position(1,8), antennas);
// place(new Position(2,5), antennas);
// place(new Position(3,7), antennas);
// antennas.print()


```

```typescript
const frequencies = {}

for (let y = 0; y < field.length; y++) {
    for (let x = 0; x < field[y].length; x++) {
        const c = field[y][x]
        if (c === '.') continue;
        // put antenna
        if (!frequencies[c]) frequencies[c] = new Positions();
        const antennas = frequencies[c];
        const p = new Position(y, x)
        place(p, antennas);
    }
}

console.log(frequencies)

globalAntinodes.print()
```
