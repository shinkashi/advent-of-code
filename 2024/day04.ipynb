```typescript
function checkDir(x:number, y:number, str: string, dx: number, dy: number) {
    let cx = x;
    let cy = y;
    let found = true;
    for (let s = 0; s < str.length; s++) {
        if (cy < 0 || cy >= l.length) { found = false; break; }
        if (cx < 0 || cx >= l[cy].length) { found = false; break; }
        if (l[cy][cx] !== str[s]) { found = false; break; }
        cx += dx;
        cy += dy;
    }
    return found;
}


checkDir(4,0,"XMAS", 0,1)
```

```typescript
function check(x:number, y:number, str: string) {
    let count = 0;

    for (let dy of [-1,0,1]) {
        for (let dx of [-1,0,1]) {
            if (dx === 0 && dy === 0) continue;
            if (checkDir(x, y, str, dx, dy)) count++;
        }
    }

    return count;
}


check(4,1,"XMAS")
```
