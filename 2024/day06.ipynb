```typescript
const lines = input.split('\n').map(ln => Array.from(ln))
lines
```

```typescript
const dirTable = {
    '^': [-1, 0],
    '<': [0, -1],
    '>': [0, 1],
    'v': [1, 0]
}


const findIntialPosition = (lines: str[]) => {
    for (let y = 0; y < lines.length; y++) {
        for (let x = 0; x < lines[y].length; x++) {
            const c = lines[y][x];
            // console.log({y,x,c});
            if (dirTable[c]) {
                const [dy, dx] = dirTable[c];
                return [y, x, dy, dx];
            }
        }
    }
}

findIntialPosition(lines)
```

```typescript
let [y, x, dy, dx] = findIntialPosition(lines)

console.log({y, x, dy, dx})
```

```typescript
const rotate = (dy, dx) => {
    const orders = [[1,0], [0,-1], [-1,0], [0,1]];
    for (let i = 0; i < orders.length; i++) {
        if (orders[i][0] == dy && orders[i][1] == dx) {
            return orders[(i + 1) % 4];
        }
    }
    throw new Error(`cannot turn right ${dy, dx} `)
}

rotate(1,0)
rotate(...rotate(1,0))
```

```typescript
const loop = (y, x, dy, dx): [number, number] => {
    let ny: number;
    let nx: number;

    while (true) {
        console.log({y, x})

        lines[y][x] = 'X';
        ny = y + dy;
        nx = x + dx;
    
        if (ny < 0 || ny >= lines.length || nx < 0 || nx >= lines[ny].length) {
            break;
        }
        
        if (lines[ny][nx] == "#") {
            [dy, dx] = rotate(dy, dx);
            continue;
        }

        y = ny;
        x = nx;
    }
    return [ny, nx];

}

loop(y, x, dy, dx)


const printField = (lines: string[][]) => {
    for (let y = 0; y < lines.length; y++) {
        console.log(lines[y].join(''));
    }
}

printField(lines)
```
